Task: Implementing Iris Flower Classification using k-Nearest Neighbors (k-NN)

```python
# Importing Libraries

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
# Representation of Dataset

iris = load_iris()
df = pd.DataFrame.from_records(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df
```

```python
# Graphical Representation
import matplotlib.pyplot as plt

data = iris.data
target = iris.target
feature_names = iris.feature_names

plt.scatter(data[:, 0], data[:, 1], c=target)
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("Scatter Plot of Iris Dataset")
plt.show()
```

```python
# Dataset Shape

df.shape
```

```python
# Dataset Splitting

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
KNN = KNeighborsClassifier(n_neighbors = 3, algorithm='auto')
KNN.fit(x_train, y_train)
```

```python
print(x_test)
prdt = KNN.predict(x_test)
print(prdt)
```

```python
# Classifier For x_test

"""for i in x_test:
    for j in prdt:
        print(f"Iris Data: {i}\n Prediction: {j}")"""
        
        
targetNames = load_iris().target_names
for data in x_test:
    prediction = KNN.predict([data])
    print(f"Iris Data: {data}\n Prediction: {targetNames[prediction[0]]}\n\n")
```

```python
# Model Accuracy

accuracy = accuracy_score(y_test, prdt)
print(f"Model Accuracy: {accuracy:.3f}\nPercentage: {accuracy*100:.2f}")
```
